```python
from sympy import *
from sympy.solvers import solve
from sympy import Symbol
from sympy.plotting import (plot,plot_parametric)
```

### Question 1

#### 1a

```python
#start code here
x = Symbol('x')
f = 1 + x*(x+1)*(x+2)*(x+3)
l = solve(f)
print("Exact form",l)
print("Approximate form",[i.evalf() for i in l])
```

#### 1b

```python
print("Expanded form:",f.expand())
```

#### 1c

```python
print("Factored form",f.factor())
```

#### 1d

```python
plot(f,(x,-3,0))
```

### Question 2

#### 2a

```python
#start code here
t = Symbol('t')
g = 9.8
hi = 6
fa = -.5*g*(t**2)+hi
l = solve(fa)
print("the textbook hits the ground at t = ", [i for i in l if i > 0])
```

#### 2b

```python
#start code here
t = Symbol('t')
g = 9.8
hi = 6
R = 2
fb = (-1*g*t)-(R*g*exp(-t/R))+(R*g)+hi/R
l = solve(fb)
print("the textbook hits the ground at t = ", [i for i in l if i > 0])
```

#### 2c

```python
#start code here
t = Symbol('t')
g = 9.8
hi = 6
R = 2
fb = (-1*g*t)-(R*g*exp(-t/R))+(R*g)+hi/R
fa = -.5*g*(t**2)+hi
fabplot = plot((fa,(t,0,3)),(fb,(t,0,3)))
```
